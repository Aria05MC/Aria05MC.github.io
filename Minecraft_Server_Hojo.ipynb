### 3. Download and Extract Mods
Downloading the zip file from your link and placing the contents in the `mods` folder.

```python
MODS_URL = "https://atleast4iq.de/Project%20Hojo/mods-4.1.0.zip"

# Force redownload and extract to ensure a fresh start
import os
print("Downloading mods...")
!wget -O mods.zip "{MODS_URL}"
print("Extracting mods...")
!unzip -o mods.zip

# --- SERVER CLEANUP & USER REQUESTED REMOVALS ---
# 'essential' added to fix 'mismatched mod list' connection error
removal_list = [
    "firstperson", "sodium", "iris", "oculus", "rubidium",
    "embeddium", "dynamiclights", "entity_model_features",
    "entity_texture_features", "notenoughanimations",
    "xaerominimap", "xaeroworldmap", "ok_zoomer",
    "mafglib", "watermedia", "freecam"
]

import os
print("Removing specified and client-side mods...")
for mod_file in os.listdir('mods'):
    if any(target in mod_file.lower() for target in removal_list):
        print(f"Removing mod: {mod_file}")
        try:
            os.remove(os.path.join('mods', mod_file))
        except:
            pass

print("\nMods cleaned and refreshed. You can now restart the server.")


# Re-download WorldEdit 7.2.15
WORLDEDIT_URL = "https://edge.forgecdn.net/files/4586/218/worldedit-mod-7.2.15.jar"
!wget -O mods/worldedit-mod-7.2.15.jar "{WORLDEDIT_URL}"

print("\n✅ SUCCESS: Folder structure flattened and mods cleaned. Start the server now!")
```

```python
import os
os.chdir('/content/minecraft_server')

def update_property(file_path, key, value):
    if not os.path.exists(file_path):
        with open(file_path, 'w') as f:
            f.write(f'{key}={value}\n')
        return

    with open(file_path, 'r') as f:
        lines = f.readlines()
    updated = False
    with open(file_path, 'w') as f:
        for line in lines:
            if line.strip().startswith(f'{key}='):
                f.write(f'{key}={value}\n')
                updated = True
            else:
                f.write(line)
        if not updated:
            f.write(f'{key}={value}\n')

# Optimization settings
update_property('server.properties', 'max-tick-time', '-1')
update_property('server.properties', 'view-distance', '3')
update_property('server.properties', 'simulation-distance', '3')

print('--- Performance Configuration Updated ---')
!grep -E "max-tick-time|view-distance|simulation-distance" server.properties
print('\nSettings applied! Please restart the START SERVER cell to apply the changes.')
```

### 6. START SERVER WITH TUNNEL
This cell starts the bore tunnel in the background and then launches the Minecraft server.

```python
import subprocess
import time
import os

# Ensure we are in the right directory
os.chdir('/content/minecraft_server')

print("🛑 NUCLEAR CLEANUP: Killing all background Java and Bore processes...")
!pkill -9 java
!pkill -9 bore
time.sleep(2)

# Start bore tunnel in the background
print("🚀 Starting Bore Tunnel...")
bore_process = subprocess.Popen(["bore", "local", "25565", "--to", "bore.pub"],
                                stdout=subprocess.PIPE,
                                stderr=subprocess.STDOUT,
                                text=True)

time.sleep(4)
for _ in range(10):
    line = bore_process.stdout.readline()
    if "listening at" in line:
        address = line.split("listening at ")[-1].strip()
        print(f"\n✅ JOIN ADDRESS: {address}")
        break

print("\n🔥 STARTING MINECRAFT SERVER IN FOREGROUND...")
print("Please wait for the 'Done!' message. Keep this cell running to keep the server alive.\n")

# Run the server directly in the foreground
!java -Xms8G -Xmx8G @user_jvm_args.txt @libraries/net/minecraftforge/forge/1.20.1-47.4.0/unix_args.txt nogui
```

```python
import psutil
import time

print("--- Checking Java Processes ---")

found = False

for proc in psutil.process_iter([
    'pid',
    'name',
    'cmdline',
    'memory_info',
    'status',
    'create_time'
]):
    try:
        name = (proc.info['name'] or '').lower()
        cmdline = ' '.join(proc.info['cmdline'] or []).lower()

        if (
            'java' in name
            or 'minecraft' in cmdline
            or 'forge' in cmdline
            or 'server.jar' in cmdline
            or 'nogui' in cmdline
        ):
            found = True

            mem = proc.info['memory_info'].rss / 1024 / 1024
            uptime = time.time() - proc.info['create_time']

            print(f"\nPID      : {proc.pid}")
            print(f"Status   : {proc.info['status']}")
            print(f"Memory   : {mem:.1f} MB")
            print(f"Uptime   : {uptime/60:.1f} min")
            print(f"Command  : {cmdline[:300]}")

    except (psutil.NoSuchProcess,
            psutil.AccessDenied,
            psutil.ZombieProcess):
        continue

if not found:
    print("\n❌ No Java/Minecraft processes found.")
```
